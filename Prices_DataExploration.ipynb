```python
# Load directly from the local file in the current directory
ds = load_dataset("parquet", data_files="stock_prices.parquet", split="train")

# Convert to pandas DataFrame
df = ds.to_pandas()

# Quick check
print("Shape:", df.shape)
```

```python
print(df.head())
```

```python
# List of the top 10 tickers from your news counts
top10_tickers = ['NVDA', 'AMZN', 'MSFT', 'TSLA', 'META',
                 'AAPL', 'INTC', 'GOOG', 'FDS', 'PLTR']

# Filter the large prices DataFrame
top10_df = df[df['symbol'].isin(top10_tickers)].copy()

# Optional: sort by date and symbol for easier time-series work
top10_df = top10_df.sort_values(['symbol', 'report_date'])

# Quick checks
print("Shape of top10_df:", top10_df.shape)
print("\nUnique tickers:", top10_df['symbol'].unique())
print("\nDate range:", top10_df['report_date'].min(), "to", top10_df['report_date'].max())
```

```python
top10_df.to_parquet("top10_stock_prices.parquet", index=False)
print("Saved to top10_stock_prices.parquet")
```

```python
top10_df.head()
```

```python
# Ensure report_date is datetime (run once if not already done)
top10_df['report_date'] = pd.to_datetime(top10_df['report_date'])

# Define the date range
start_date = pd.to_datetime('2025-03-12')
end_date   = pd.to_datetime('2026-03-15')

# Filter rows within the date range (inclusive)
narrow_df = top10_df[
    (top10_df['report_date'] >= start_date) &
    (top10_df['report_date'] <= end_date)
].copy()

# Sort for consistency
narrow_df = narrow_df.sort_values(['symbol', 'report_date'])

# Quick checks
print("Shape of narrow_df:", narrow_df.shape)
print("Date range:", narrow_df['report_date'].min(), "to", narrow_df['report_date'].max())
print("Unique tickers:", sorted(narrow_df['symbol'].unique()))
print("Expected ~2510 rows → actual:", len(narrow_df))
```

```python
narrow_df.to_parquet("top10_prices_recent.parquet", index=False)
print("Saved to top10_prices_recent.parquet")
```
