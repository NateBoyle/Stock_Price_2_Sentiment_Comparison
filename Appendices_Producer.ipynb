```python
# Company names dictionary
company_names = {
    'NVDA': 'NVIDIA Corporation',
    'META': 'Meta Platforms, Inc.',
    'PLTR': 'Palantir Technologies Inc.',
    'INTC': 'Intel Corporation',
    'AAPL': 'Apple Inc.',
    'MSFT': 'Microsoft Corporation',
    'AMZN': 'Amazon.com, Inc.',
    'FDS': 'FactSet Research Systems Inc.',
    'TSLA': 'Tesla, Inc.',
    'GOOG': 'Alphabet Inc. (Google)'
}

doc = Document()
doc.add_heading('Appendix A: Savitzky-Golay Smoothed Dual-Axis Time Series Graphs', level=1)

tickers = ['NVDA', 'META', 'PLTR', 'INTC', 'AAPL', 'MSFT', 'AMZN', 'FDS', 'TSLA', 'GOOG']

for ticker in tickers:
    company = company_names.get(ticker, ticker)
    short_name = company.split()[0]   # e.g., "Apple", "NVIDIA", etc.

    # Exact title format from your document
    doc.add_heading(f'{ticker} ({short_name}) Savitzky-Golay Graphs', level=2)

    # FinBert vs VADER header
    doc.add_paragraph('FinBert                                                  VADER', style='Heading 3')

    for ret_type in ['daily_return', 'intraday_return', 'next_day_return']:
        nice_name = ret_type.replace('_', ' ').title()
        doc.add_heading(nice_name, level=3)

        # FinBert Savgol graph
        finbert_file = f"{ticker.lower()}_finbert_vs_{ret_type}_savgol.png"
        if os.path.exists(finbert_file):
            doc.add_picture(finbert_file, width=Inches(3.2))
        else:
            doc.add_paragraph(f"[Missing: {finbert_file}]")

        doc.add_paragraph(" ")  # spacing between images

        # VADER Savgol graph
        vader_file = f"{ticker.lower()}_vader_vs_{ret_type}_savgol.png"
        if os.path.exists(vader_file):
            doc.add_picture(vader_file, width=Inches(3.2))
        else:
            doc.add_paragraph(f"[Missing: {vader_file}]")

        doc.add_paragraph("\n")   # extra space before next return type

    doc.add_page_break()

doc.save('Appendix_A_Savitzky_Golay_Graphs.docx')
print("✅ Appendix A created successfully!")
```
